4. Las columnas que son string ver cuales son los valores únicos de cada columna

```python
def contar_valores_unicos(base, lista_columnas):
    for col in lista_columnas:
        columna = pd.unique(base[col]).shape[0]
        print('La columna {} tiene {} valores unicos'.format(str(col), str(columna)))

lista_columnas = ['mes_inicio', 'fecha_inicio', 'delito', 'fiscalia', 'agencia', 'unidad_investigacion',   
                  'categoria_delito', 'calle_hechos', 'calle_hechos2', 'colonia_hechos', 'alcaldia_hechos',          
                  'competencia']
contar_valores_unicos(base, lista_columnas=lista_columnas)
```

Podemos ver que aparte de la fecha de inicio, calle_hechos y calle_hechos2 las columnas tienen pocos valores únicos, por lo tanto podemos deducir que están bien catalogados

### Determinar los delitos que van al alza o a la baja

```python
base_hechos = base.groupby(['ao_hechos' ,'delito']).count().reset_index()[['ao_hechos', 'delito', 'categoria_delito']]

base_anterior = base_hechos[(base_hechos['ao_hechos'] >= 2018) & (base_hechos['ao_hechos'] < 2020)].groupby(['delito'])[['categoria_delito']].sum().reset_index()
base_posterior = base_hechos[(base_hechos['ao_hechos'] <= 2021) & (base_hechos['ao_hechos'] >= 2020)].groupby(['delito'])[['categoria_delito']].sum().reset_index()
base_anterior.columns = ['delito', 'delito_2018_2019']
base_posterior.columns = ['delito', 'delito_2020_2021']
base_hechos = pd.merge(base_anterior, base_posterior, on=['delito'])
base_hechos['ratio'] =  base_hechos['delito_2020_2021']/base_hechos['delito_2018_2019']
base_hechos = base_hechos.sort_values(['ratio'])
```

### Determinar la alcaldia que tiene mas delitos y la de menos

```python
base_alcaldia = base.groupby(['alcaldia_hechos']).count().reset_index()[['alcaldia_hechos','ao_hechos']]
base_alcaldia.sort_values(['ao_hechos'], ascending=False)
```
